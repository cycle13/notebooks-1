### 1. 问题
用了python的[`xarray`](http://xarray.pydata.org/en/stable/)包后，特别喜欢它的数据接口，直接就可以读取自描述的[NetCDF](https://www.unidata.ucar.edu/software/netcdf/)（nc）文件，并且组织成一个`Dataset`。例如打开[`xarray`](http://xarray.pydata.org/en/stable/)自带的教学nc文件可以得到如下的`Dataset`：

```python
import xarray as xr

dset = xr.tutorial.open_dataset('air_temperature')
dset
```

可以很清楚地看到nc文件的结构、变量、坐标、meta信息等等。

我以前是[GrADS/OpenGrADS](http://www.opengrads.org/)的重度用户，我也习惯使用[GrADS/OpenGrADS](http://www.opengrads.org/)和配套的二进制数据，通常是一个二进制文件配一个.ctl（control，类似于nc的header信息）文件进行信息描述。所以现在开始学习python以后希望能用python直接解析ctl文件信息然后读取相应的二进制文件数据。解析ctl文件并不难，但是我希望能和[`xarray`](http://xarray.pydata.org/en/stable/)无缝连接，读取完数据返回一个`Dataset`，最难的还不是这个，而是使用[`dask`](https://dask.org/)工具，做到数据在内存中的缓存和并行。

---

### 2. 解决方法
首先，编写了一个[`CtlDescriptor`]()类来代表.ctl文件，使用如下：

```python
import sys
sys.path.append('d:/Python/MITgcm/')
from utils.IOUtils import CtlDescriptor

ctl = CtlDescriptor(file='D:/Data/ERAInterim/Keff/PV/PV.ctl')

print(ctl)
```

然后用这个类来构建一个`open_CtlDataset`函数（放在[IOUtils.py]()里面），该函数和[`xarray.open_dataset`](http://xarray.pydata.org/en/stable/generated/xarray.open_dataset.html)一样返回`Dataset`：

```python
from utils.IOUtils import open_CtlDataset

dset = open_CtlDataset('D:/Data/ERAInterim/Keff/PV/PV.ctl')

dset
```

花了点时间去编写[`dask`](https://dask.org/)相关代码，所以这里返回的`Dataset`同样支持[`dask`](https://dask.org/)的`chunk`功能。这样python可以完美读取[GrADS/OpenGrADS](http://www.opengrads.org/)的ctl二进制文件，并且与[`xarray`](http://xarray.pydata.org/en/stable/)的功能无缝配合。比如读取变量并画图：

```python
dset.pv[0].plot()
```
